### **PROBLEM STATEMENT**

Olist, a well-known e-commerce platform in Brazil, is dealing with a **serious challenge when it comes to customer retention**. Data from 2016 to 2018 shows that **97% of customers only make one purchase and do not come back, leaving just 3% as repeat buyers**. This low retention rate has a significant impact on Olist's revenue and growth potential.

Based on the provided dataset, the customer proportions by the number of purchases are:
  - **One-time Buyers** (Single Purchase): **90,859 customers** (97%)
  - **Repeat Buyers** (Multiple Purchases): **2,825 customers** (3%)
  
According to [Olist Profit](https://medium.com/@fercanc/how-can-olist-increase-its-profit-ef58e3f274e#:~:text=P%26L%20Rules,each%20order%20delivered.), Olist earns a 10% commission on the product price (excluding freight) for each order delivered. From the dataset, we calculated that the **average revenue per customer’s order is approximately BRL 14.44**, derived from this 10% commission. Additionally, our calculations show that **repeat buyers make an average of about 2 purchases each**.

Therefore, Olist's revenue from commissions between 2016 and 2018 can be calculated as follows:
1. **Revenue from One-time Buyers:**
    - **Total Number of One-time Buyers** x **Commission per Order** = 90,859 customers x BRL 14.44 = BRL 1,312,003

2. **Revenue from Repeat Buyers:**
    - **Total Orders by Repeat Buyers:** 
      - **Total Number of Repeat Buyers** x **Average Purchases per Repeat Buyer** = 2,825 customers x 2 purchases = 5,650 orders
    - **Total Revenue from Repeat Buyers:**
      - **Total Orders by Repeat Buyers** x **Commission per Order** = 5,650 orders x BRL 14.44 = BRL 81,586

3. **Total Revenue from All Customers:**
    - **Total Revenue from One-time Buyers** + **Total Revenue from Repeat Buyers** = BRL 1,312,003 + BRL 81,586 = **BRL 1,393,589**

The code-generated calculation confirms that Olist's revenue is also approximately BRL 1.39 million, based on summing all commissions (10% of the product prices) from all ordered products. This verifies that the previous information is accurate and aligns with the dataset. This information can be used to estimate the potential revenue growth for each percentage increase in the retention rate.

**Olist’s revenue of approximately BRL 1.39 million from 2016 to 2018 represents only a tiny fraction of the overall Brazilian e-commerce market**. ### **ANALYTICAL APPROACH (MODELS AND EVALUATION METRICS)**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('df_olist_clean.csv')
df
```

## **BUSINESS PROBLEM AND TARGET CALCULATION**

```python
# total number of customers
total_cust = df['customer_unique_id'].nunique()



# customer distribution based on the number of purchases and their percentage
customer_order_counts = df.groupby('customer_unique_id')['order_id'].nunique()

buyer_classification = []
for count in customer_order_counts:
    if count > 1:
        buyer_classification.append('Repeated Buyer')
    else:
        buyer_classification.append('One-Time Buyer')

buyer_classification_series = pd.Series(buyer_classification, index=customer_order_counts.index)

distribution = buyer_classification_series.value_counts()
distribution_percentage = ((distribution / distribution.sum()) * 100).round(1)

distribution_df = pd.DataFrame({
    'Buyer Type': distribution.index,
    'Count': distribution.values,
    'Percentage': distribution_percentage.values
})



# average number of purchases per repeat buyer
repeat_buyers = customer_order_counts[customer_order_counts > 1]

total_repeat_orders = repeat_buyers.sum()

num_repeat_buyers = len(repeat_buyers)

average_repeat_orders_per_buyer = total_repeat_orders / num_repeat_buyers



# Olist commission (10% of the product's price)
olist_cut_percentage = 0.10 

df['olist_commission'] = df['price'] * olist_cut_percentage

commission_df = df[['customer_unique_id', 'order_id', 'price', 'olist_commission']]



# average revenue per customer's order (commission from 10% of the product's price)
order_revenue = df.groupby('order_id')['olist_commission'].sum()

average_revenue_per_order = order_revenue.mean()



# total revenue from all customers
total_revenue = df['olist_commission'].sum()



print(f"Total Customers: {total_cust} customers")
display(distribution_df)
print(f"Average Number of Purchases per Repeat Buyer: {average_repeat_orders_per_buyer:.0f} purchases")
display(commission_df.head(6))
print(f"Average Revenue per Customer's Order: BRL {average_revenue_per_order:.2f}")
print(f"\nTotal Revenue from All Customers: BRL {total_revenue:.2f}")
```
